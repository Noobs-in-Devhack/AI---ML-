```python
import numpy as np
# Data Manipulation & Wrangling
import pandas as pd
# Basic Visualization - Matplotlib : Bar,Line,Scatter,Histogram,Pie etc.
import matplotlib.pyplot as plt
# Complex Visualization -
import seaborn as sns

# Disable Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
auto_price = pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Automobile_price_data__Raw_.csv")
auto_price.head()
```

```python
# replace '?' from all columns by np.nan
auto_df = auto_price.replace('?', np.nan)
```

```python
auto_df.isnull().sum()
```

```python
auto_df.drop(['symboling','normalized-losses'], axis = 1, inplace = True)
```

```python
auto_df.dropna(inplace = True)    # Drop all rows with missing values
```

```python
cols = ['bore','stroke','horsepower','peak-rpm','price']
# Function to change columns to numeric datatype
auto_df[cols] = auto_df[cols].apply(pd.to_numeric)
```

#### Line Plot
    * Univariate or Bivariate (Relationship Plot).
    * Trend Analysis (Time Series) - Univariate Analysis.

```python
auto_df['price'].sort_values()      # sorting price by values in an Ascending Order
```

```python
data = auto_df['price'].sort_values().unique()
data
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (8,6))
plt.plot(data)
plt.show()
```

```python
auto_df.sort_values(by = 'price')
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (8,6))
plt.plot(auto_df.sort_values(by = 'price')['price'], auto_df.sort_values(by = 'price')['city-mpg'], color = 'darkviolet')
plt.show()
```

#### Bar Plot
    * Bar chart used by Categorical, Nominal, Categorical vs Value Counts/Numerical/Continious
    * Bar plot is used for Univariate and Bivariate Analysis.

```python
data = auto_df['make'].value_counts()
data.index
```

```python
plt.figure(figsize = (8,6))
plt.bar(x = data.index, height = data.values, color = 'blue')
plt.xticks(rotation = 90)
plt.show()
```

Histogram

    * Continious Data - Study the spread/distribution of continious data.
    * Univariate Analysis.

```python
plt.figure(figsize = (8,5))
plt.hist(auto_df['price'], bins = 15)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title("Histogram Plot - Price")
plt.show()
```

#### Scatter Plot
    * Scatter Plots is plotted between numerical values.
    * It shows relationship b/w x & y plots
    * Bivariate Analysis.  

```python
plt.figure(figsize = (9,6))
sns.set_style('whitegrid')
sns.scatterplot(x = auto_df['city-mpg'], y = auto_df['price'], color = 'deeppink', marker = 's', alpha = 0.7)
```

#### Pie-Chart

```python
auto_df['body-style'].value_counts()
```

```python
plt.figure(figsize = (6,6))
plt.pie(auto_df['body-style'].value_counts(), labels = auto_df['body-style'].value_counts().index,
        autopct = '%0.2f%%')
plt.show()
```

#### Distribution Plot
    * Histogram + KDE (Kernel Density Estimator)
    * Probability Distribution Function

```python
sns.distplot(auto_df['engine-size'], bins = 20, color = 'blue')
plt.show()
```

```python
sns.distplot(auto_df['horsepower'], bins = 20, color = 'blue', hist = False)
plt.show()
```

#### Boxplot
    * Distribution of continious data
    * It shows statistical distribution like Quartiles (Q1,Q2,Q3) and Median (50% or Q2).
    * Plot will be useful in showing outliers.

```python
plt.figure(figsize = (7,3))
sns.boxplot(auto_df['peak-rpm'], orient = 'h', color = 'violet')
```
