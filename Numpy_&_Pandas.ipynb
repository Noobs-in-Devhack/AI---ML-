#### Types of Arrays - 2D, 3-D

```python
# 2D Array - Matrix
arr_2 = np.array([[10, 20, 30, 40, 50, 60, 70],
                  [21, 22, 23, 25, 27, 28, 32],
                  [31, 35, 36, 38, 43, 47, 51]])
print(arr_2)
```

```python
print(arr_2.shape)      # returns shape of numpy array - num of rows, num of columns
print(arr_2.ndim)       # num of dimentionality
print(arr_2.size)       # num of elements in an array
```

```python
# 3D Array - Collection of Matrices
arr_3 = np.array([[[10, 20, 30, 40, 50, 60, 70],
                  [21, 22, 23, 25, 27, 28, 32],
                  [31, 35, 36, 38, 43, 47, 51]],

                 [[11, 21, 31, 40, 52, 63, 75],
                  [21, 22, 23, 29, 27, 28, 32],
                  [31, 35, 38, 38, 43, 47, 51]]])
print(arr_3)
```

```python
print(arr_3.shape)      # returns shape of numpy array - num of rows, num of columns
print(arr_3.ndim)       # num of dimentionality
print(arr_3.size)       # num of elements in an array
```

#### Pandas
    * Pandas are useful python package, reading various file formats.
    * Data Wrangling/Manipulation format
    * Pandas is good for data analysis and data visualization.
    * DataFrame is 2D dataset respresented as rows and columns in 2D format just like spreadsheet.

```python
import pandas as pd
```

```python
auto_price = pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Automobile_price_data__Raw_.csv")
```

```python
auto_price.head()    # show first five rows of dataframe
```

```python
auto_price.tail()    # show last five rows of dataframe
```

```python
auto_price.shape    # shape (rows, columns)
```

```python
auto_price.columns     # returns list of columns
```

#### Data Manipulation

```python
auto_price.info()
```

```python
# replace '?' by NaN
auto_price.replace('?', np.nan, inplace = True)
```

```python
auto_price.isnull().sum()   # count of missing values for each column
```

#### Slicing & Indexing DataFrame

```python
# loc = slicing rows & columns by label names
auto_price.loc[:, :]
```

```python
# loc : slicing rows start index = 1, end index = 7
auto_price.loc[1:8, :]
```

```python
# loc : slice rows and columns by label names
# column labels : start = 'make', end index : 'engine-type'
auto_price.loc[:, 'make':'engine-type']
```

```python
# loc[list of rows, list of columns]
auto_price.loc[[2,4,6,8,10,12,14,16], ['make','fuel-type','num-of-doors','body-style','engine-size','price']]
```

```python
# iloc = index location, slicing by index
auto_price.iloc[1:10, :]
```

```python
# iloc = index location, slicing by index
auto_price.iloc[1:10, 2:11]
```

#### Select Columns

```python
auto_price['price']    # select column by column names
```

```python
auto_price[['make','body-style','fuel-type','price']]    # select multiple columns by individual columnn names
```

#### Filter
    * To filter rows by boolean condition

```python
auto_price['make'] == 'audi'
```

```python
auto_price[auto_price['make'] == 'audi']
```

```python
auto_price[auto_price['highway-mpg'] >= 45]
```

```python
auto_price[(auto_price['highway-mpg'] >= 40) & (auto_price['body-style'] == 'sedan')]
```

```python
# Display distinct or unique values. (Apply it on categorical columns)
auto_price['make'].unique()
```

```python
auto_price['make'].value_counts()     # Frequency of Categorical Column
```

#### Drop()
    * Dropping Rows & Columns

```python
# Drops columns by using indexes
auto_price.drop(['symboling','normalized-losses'], axis = 1, inplace = True)
```

```python
auto_price.head()
```

```python
# Drops rows by using indexes
auto_price.drop([0, 1, 2, 3, 4, 5, 6, 7, 8], axis = 0)
```

#### Statistical Methods

```python
# Typecasting string to float data type
auto_price['price'] = auto_price['price'].astype('float')
```

```python
auto_price['horsepower'] = auto_price['horsepower'].astype('float')
```

```python
np.round(auto_price['price'].mean())
```

```python
auto_price['price'].median()
```

```python
auto_price['make'].mode()    # most-frequent value
```
